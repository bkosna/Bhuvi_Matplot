```python
dep = pd.read_csv('cces_example_data.csv')
dep.head()
```

```python
fig, axal = plt.subplots(1, 2, figsize=(6, 3))

axal[0].set_title("Age Histogram")
axal[1].set_title("Political Identity")

plt.show()
```

```python
evaluation_counts = dep['Evaluation'].value_counts()

plt.figure(figsize=(4, 2.5))
evaluation_counts.plot(kind='bar', color='skyblue')
plt.title('Evaluation Responses')
plt.xlabel('Evaluation Categories')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(4, 2.5))
plt.scatter(dep['ConMediaUse'], dep['ImmiPhobia'], alpha=0.6)
plt.title('Scatter Plot of ConMediaUse vs. ImmiPhobia')
plt.xlabel('ConMediaUse')
plt.ylabel('ImmiPhobia')
plt.grid(True)
plt.show()
```
